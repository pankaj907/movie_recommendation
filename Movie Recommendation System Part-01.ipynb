4     False  False


movies2.isnull().sum()

This command adds up the boolean values in each column.
The sum() method treats True as 1 and False as 0.
It then calculates the sum for each column, telling you the total number of missing values per column. 

Example:
Using the same DataFrame as above, movies2.isnull().sum() would return:

```python
import ast
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i["name"])
    return L
```

```python
import ast
def convert3(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            L.append(i["name"])
            counter=counter+1
        else:
            break
        
    return L
```

```python
import ast
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i["job"]=="Director":
            L.append(i["name"])
            break
    return L
```
