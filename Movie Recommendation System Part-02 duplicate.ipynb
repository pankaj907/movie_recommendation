```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

""" import nltk: This line imports the Natural Language Toolkit (NLTK) library, which is a comprehensive library for working with human language data.

from nltk.stem.porter import PorterStemmer: This line specifically imports the PorterStemmer class from the nltk.stem.porter module within the NLTK library. The Porter Stemmer is a well-known algorithm for reducing words to their root or base form (stemming)."""



```python
print(ps.stem("dance"))
```

```python
def stem(text):
    y=[]

    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

"""text.split(): The input text is split into individual words based on whitespace, creating a list of words (tokens).
for i in ...: The code iterates through each word i in the list of tokens.


ps.stem(i): For each word i, the stem() method of an object named ps is called. This ps object is expected to be an instance of a stemmer, such as PorterStemmer from the nltk library, which reduces the word to its root form (stem).


y.append(...): The stemmed word is appended to the list y.   

return " ".join(y)
Finally, the elements (stemmed words) in 
the list y are joined back together into a single string,
with each word separated by a space, and 
this resulting string is returned."""

```python
new_df["tags"].apply(stem)
```

```python
new_df["tags"]=new_df["tags"].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors)
```

"""The cosine_similarity function calculates the cosine similarity between samples 
represented as vectors.Cosine similarity measures the cosine of the angle between two non-zero vectors.

A value of 1 indicates that the vectors are pointing in the same direction (maximum similarity),
0 indicates orthogonality (no similarity), 

and -1 indicates that they are pointing in opposite directions (maximum dissimilarity)."""

```python
similarity=cosine_similarity(vectors)
```

"""enumerate(similarity[0]) = returning both the index and the value of each item"""

```python
def recommend(movie1):
    movie_index=new_df[new_df["title"]==movie1].index[0]
    distances=similarity[movie_index]
    movie_list=sorted(list(enumerate(distances)),reverse= True,key=lambda x:x[1])[1:6]
    for i in movie_list:
         print(new_df.iloc[i[0]].title)
```

```python
recommend("Batman")
```
